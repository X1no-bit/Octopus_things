```python
import pandas as pd
from sklearn.model_selection import train_test_split
ratings = pd.read_csv('kinopoisk-top250.csv')
if 'rating' in ratings.columns and 'movie' in ratings.columns and 'rating_ball' in ratings.columns:
    user_ids = ratings['rating'].unique()
    movie_ids = ratings['movie'].unique()
    user2idx = {u: i for i, u in enumerate(user_ids)}
    movie2idx = {m: i for i, m in enumerate(movie_ids)}
    ratings['rating'] = ratings['rating'].map(user2idx)  
    ratings['movie'] = ratings['movie'].map(movie2idx)  
    ratings = ratings[['rating', 'movie', 'rating_ball']]
    train, test = train_test_split(ratings, test_size=0.2)
    print("Обучающая выборка:")
    print(train.head())
    print("Тестовая выборка:")
    print(test.head())

```
