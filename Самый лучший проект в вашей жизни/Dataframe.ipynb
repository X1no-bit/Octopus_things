```python
import pandas as pd

def filter_matches(df, your_team_picks, enemy_team_picks):
    """
    Фильтрует матчи, где одна команда содержит ВСЕХ героев из your_team_picks,
    а противоположная - ВСЕХ героев из enemy_team_picks
    
    Параметры:
    df (DataFrame): Исходные данные матчей
    your_team_picks (list): ID героев вашей команды (1-5 элементов)
    enemy_team_picks (list): ID героев противника (1-5 элементов)
    
    Возвращает:
    DataFrame с отфильтрованными матчами
    """
    # Преобразуем списки в множества
    your_set = set(your_team_picks)
    enemy_set = set(enemy_team_picks)
    
    # Функция проверки полного вхождения героев
    def contains_all_heroes(row, team_side, heroes):
        team_heroes = {row[f'{team_side}_hero_{i}'] for i in range(1, 6)}
        return heroes.issubset(team_heroes)
    
    # Проверяем два возможных сценария
    mask_scenario1 = df.apply(
        lambda x: contains_all_heroes(x, 'radiant', your_set) 
        and contains_all_heroes(x, 'dire', enemy_set),
        axis=1
    )
    
    mask_scenario2 = df.apply(
        lambda x: contains_all_heroes(x, 'dire', your_set) 
        and contains_all_heroes(x, 'radiant', enemy_set),
        axis=1
    )
    
    # Комбинируем маски
    result_df = df[mask_scenario1 | mask_scenario2]
    
    return result_df.reset_index(drop=True)
```

```python
import pandas as pd
from collections import defaultdict

def precompute_hero_statistics(df):
    """Предварительно вычисляет статистику героев: синергии, контрпики и винрейты."""
   
    required_columns = [f'{team}_hero_{i}' for team in ['radiant', 'dire'] for i in range(1, 6)]
    required_columns.append('radiant_win')
    
    missing_cols = [col for col in required_columns if col not in df.columns]
    if missing_cols:
        raise ValueError(f"В данных отсутствуют обязательные колонки: {missing_cols}")

    hero_pairs_synergy = defaultdict(lambda: defaultdict(lambda: [0, 0]))  # [wins, total]
    hero_pairs_counter = defaultdict(lambda: defaultdict(lambda: [0, 0]))
    hero_win_rates = defaultdict(lambda: [0, 0])  # [wins, total]
    
    all_heroes = set()
    for i in range(1, 6):
        all_heroes.update(df[f'radiant_hero_{i}'].dropna().unique())
        all_heroes.update(df[f'dire_hero_{i}'].dropna().unique())
    
 
    for _, row in df.iterrows():
        radiant_heroes = [row[f'radiant_hero_{i}'] for i in range(1, 6) if not pd.isna(row[f'radiant_hero_{i}'])]
        dire_heroes = [row[f'dire_hero_{i}'] for i in range(1, 6) if not pd.isna(row[f'dire_hero_{i}'])]
        radiant_win = row['radiant_win']
        
        for hero in radiant_heroes:
            hero_win_rates[hero][1] += 1
            hero_win_rates[hero][0] += 1 if radiant_win else 0
        for hero in dire_heroes:
            hero_win_rates[hero][1] += 1
            hero_win_rates[hero][0] += 0 if radiant_win else 1
        

        for team in [radiant_heroes, dire_heroes]:
            for i in range(len(team)):
                for j in range(i + 1, len(team)):
                    h1, h2 = team[i], team[j]
                    hero_pairs_synergy[h1][h2][1] += 1
                    hero_pairs_synergy[h2][h1][1] += 1
                    if (team is radiant_heroes and radiant_win) or (team is dire_heroes and not radiant_win):
                        hero_pairs_synergy[h1][h2][0] += 1
                        hero_pairs_synergy[h2][h1][0] += 1
        
        for r_hero in radiant_heroes:
            for d_hero in dire_heroes:
                hero_pairs_counter[r_hero][d_hero][1] += 1
                hero_pairs_counter[d_hero][r_hero][1] += 1
                if radiant_win:
                    hero_pairs_counter[r_hero][d_hero][0] += 1
                else:
                    hero_pairs_counter[d_hero][r_hero][0] += 1
  
    synergy_rates = {}
    for h1, partners in hero_pairs_synergy.items():
        synergy_rates[h1] = {}
        for h2, (wins, total) in partners.items():
            synergy_rates[h1][h2] = wins / total if total > 0 else 0.5
    
    counter_rates = {}
    for h1, opponents in hero_pairs_counter.items():
        counter_rates[h1] = {}
        for h2, (wins, total) in opponents.items():
            counter_rates[h1][h2] = wins / total if total > 0 else 0.5
    
    win_rates = {hero: wins / total for hero, (wins, total) in hero_win_rates.items()}
    
    return {
        'synergy_rates': synergy_rates,
        'counter_rates': counter_rates,
        'win_rates': win_rates,
        'all_heroes': list(all_heroes)
    }

def get_hero_recommendations(
    player_side,          
    team_heroes,         
    enemy_heroes,         
    banned_heroes,        
    stats,               
    preferred_heroes=[],  
    n_picks=5,          
    n_bans=5,            
    synergy_weight=0.4,   
    counter_weight=0.4,   
    win_rate_weight=0.2  
):
    """Рекомендует героев для пика и бана на основе статистики матчей."""
    available_heroes = [
        h for h in stats['all_heroes'] 
        if h not in banned_heroes + team_heroes + enemy_heroes
    ]
    pick_scores = []
    for hero in available_heroes:
        synergy_score = 0
        if team_heroes:
            for teammate in team_heroes:
                if teammate in stats['synergy_rates'].get(hero, {}):
                    synergy_score += stats['synergy_rates'][hero][teammate]
            synergy_score /= len(team_heroes) if team_heroes else 1
        counter_score = 0
        if enemy_heroes:
            for enemy in enemy_heroes:
                if enemy in stats['counter_rates'].get(hero, {}):
                    counter_score += stats['counter_rates'][hero][enemy]
            counter_score /= len(enemy_heroes) if enemy_heroes else 1
        win_score = stats['win_rates'].get(hero, 0.5)
        preference_bonus = 0.1 if hero in preferred_heroes else 0
        total_score = (
            synergy_weight * synergy_score +
            counter_weight * counter_score +
            win_rate_weight * win_score +
            preference_bonus
        )
        pick_scores.append((hero, total_score))
    pick_scores.sort(key=lambda x: x[1], reverse=True)
    recommended_picks = [hero for hero, score in pick_scores[:n_picks]]
    ban_candidates = [h for h in stats['all_heroes'] if h not in banned_heroes]
    ban_scores = []
    
    for hero in ban_candidates:
        danger_score = 0
        if team_heroes:
            for teammate in team_heroes:
                if hero in stats['counter_rates'].get(teammate, {}):
                    danger_score += 1 - stats['counter_rates'][teammate][hero]  
            danger_score /= len(team_heroes) if team_heroes else 1
        win_score = stats['win_rates'].get(hero, 0.5)
        
        ban_scores.append((hero, danger_score * 0.7 + win_score * 0.3))
    
    ban_scores.sort(key=lambda x: x[1], reverse=True)
    recommended_bans = [hero for hero, score in ban_scores[:n_bans]]
    
    return {
        'recommended_picks': recommended_picks,
        'recommended_bans': recommended_bans,
        'pick_scores': pick_scores[:n_picks],
        'ban_scores': ban_scores[:n_bans]
    }

if __name__ == "__main__":
    try:
        df = combined_df  

        print("Вычисление статистики героев...")
        stats = precompute_hero_statistics(df)
        print("Статистика успешно вычислена!")
        
        print("\nГенерация рекомендаций...")
        recommendations = get_hero_recommendations(
            player_side='radiant',
            team_heroes=[11, 32],    # Shadow Fiend и Jakiro
            enemy_heroes=[45, 23],    #Pudge и Phantom Assassin
            banned_heroes=[18, 53],   # Invoker и Anti-Mage
            stats=stats,
            preferred_heroes=[74, 20] #Templar Assassin и Brewmaster
        )
        
        print("\nТоп рекомендуемых пиков:")
        for i, (hero, score) in enumerate(recommendations['pick_scores'], 1):
            print(f"{i}. Герой {hero}: {score:.2%}")
            
        print("\nТоп рекомендуемых банов:")
        for i, (hero, score) in enumerate(recommendations['ban_scores'], 1):
            print(f"{i}. Герой {hero}: {score:.2%}")
            
    except Exception as e:
        print(f"Ошибка: {str(e)}")
```
